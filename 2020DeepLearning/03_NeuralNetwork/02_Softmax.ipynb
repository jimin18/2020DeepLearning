```
import numpy as np
```

```
a = np.array([0.3, 0.9, 5.0])
```

```
a = np.array([0.3, -0.9, 5.0])
```

```
a
```

```
def softmax(a): 
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a) 
  y = exp_a / sum_exp_a 
  return y
```

```
y = softmax(a)
```

```
print(y)
```

```
np.sum(y)
```

Numerically stable Softmax

```
a = np.array([1010, 1000, 990])
```

```
y = softmax(a)
```

```
print(y)
```

```
def stable_softmax(a):
  c = np.max(a)
  exp_a = np.exp(a-c)
  sum_exp_a = sum(np.exp(a-c))
  y = exp_a/sum_exp_a
  return y
```

```
y = stable_softmax(a)
```

```
print(y)
```

```
np.sum(y)
```
