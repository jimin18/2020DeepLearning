```
import numpy as np
```

Mean Square Error (MSE)

```
def MSE(y, t):
    loss = 0.5 * np.sum((y-t)**2)
    return loss
```

```
t = [ 0, 10, 0, 0, 0, 0, 0, 0, 0, 0]

y1 = [0.3, 0.02, 0.3, 0.0, 0.01, 0.05, 0.0, 0.12, 0.14, 0.04]
y2 = [0.2, 0.01, 0.0, 0.5, 0.02, 0.01, 0.0, 0.15, 0.1, 0.02]
y3 = [0.2, 0.03, 0.4, 0.2, 0.03, 0.07, 0.7, 0.13, 0.2, 0.03]
```

```
loss1 = MSE(np.array(t), np.array(y1))
loss2 = MSE(np.array(t), np.array(y2))
loss3 = MSE(np.array(t), np.array(y3))
```

```
print(loss1)
print(loss2)
print(loss3)
```

$$ loss =  - \Sigma t * log(y) $$

```
def CEE(y, t):
    epsilon = 0.00000001
    v_temp = t * np.log(y + epsilon)
    loss = -np.sum(v_temp)
    return loss
```

```
l1 = CEE(np.array(t), np.array(y1))
print(l1)
```

```
l2 = CEE(np.array(t), np.array(y2))
print(l2)
```
