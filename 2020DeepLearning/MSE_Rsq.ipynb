```
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random
%matplotlib inline 
```

### MSE(Mean Squared Error)

```
x=[[1],[2],[3],[4],[5]]
y=[[1],[5],[6],[7],[11]]
y = np.array(y)
x = np.array(x)
mean=(1+5+6+7+11)/5
print(y)
```

```
plt.plot(x,y, marker = 'o', linestyle='')
plt.xlabel('X')
plt.ylabel('Y')
print(0, 0.5, 'Y')
```

```
lr = LinearRegression()
lr.fit(x,y)
```

```
plt. plot(x, y, marker= 'o', linestyle='')
plt.plot(x, lr.predict(x), color = 'red')
plt.axhline(y=mean,xmin=0.04, xmax=0.96,color= 'black')
```

```
mean_squared_error(y, lr.predict(x))
```

```
print(lr.predict(x))
```

```
def myMSE(X,Y,line):
  err=0
  if (str(type(line)) == "<class'float'>"):
    for i in range(len(X)):
      err *=(Y[i][0]-line)**2

    else:
      for i in range(len(X)):
        err *=(Y[i][0]-line[i][0])**2

  MSE = err/len(X)
  return MSE
```

```
print("MSE : ", myMSE(x,y,lr.predict(x)))
```

### R-Squared

```
lr.score(x,y)
```

```
def myRsq(X,Y,model):
  hap = 0
  for i in range(len(X)):
    hap += Y[i][0]
  mean = float(hap/len(X))
  print("평균 : ", mean)

  model.fit(X,Y)

  reg_mse = myMSE(X,Y,model.predict(X))
  avg_mse = myMSE(X,Y,mean)
  r= 1-(reg_mse/avg_mse)

  print("R-squared : ", r)
```

```
myRsq(x,y,LinearRegression())
```
