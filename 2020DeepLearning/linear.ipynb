```
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x_train = [1,2,3,4,5]
y_train = [4,8,12,16,20]
#y=4x
signal_length = len(x_train)
y_noise = np.random.normal(0,1,signal_length)

y_train = y_train + y_noise
```

```
plt.plot(x_train, y_train,'.-')
plt.grid()
```

```
useRandom = False
```

```
if useRandom:
  w = tf.Variable(tf.random_normal([1]), name='weight')
  b = tf.Variable(tf.random_normal([1]), name="bias")

else:
  w0 = 5.0;
  b0 = 3.0;

  w = tf.Variable(w0*tf.ones([1]), name='weight')
  b = tf.Variable(b0*tf.ones([1]), name='bias')
```

```
hypothesis=x_train * w + b
```

```
loss = tf.reduce_mean(tf.square(hypothesis - y_train))
```

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(loss)
```

```
sess = tf.Session()
```

```
sess.run(tf.global_variables_initializer())
```

```
nb_epoch = 1001
vloss=[]

for step in range(nb_epoch):
  sess.run(train)
  loss1 = sess.run(loss)
  vloss.append(loss1)

  if step % 50 == 0:
    w1 = sess.run(w)[0]
    b1 = sess.run(b)[0]

    print(step,'\t', loss1, '\t', 1, '\t', b1)
```

```
plt.plot(vloss[:20],'.-')
plt.xlabel('epoch')
plt.ylabel('loss')
```

```
nb_epoch = 1001
vb=[]

for step in range(nb_epoch):
  sess.run(train)
  b1 = sess.run(b)
  vb.append(b1)

  if step % 50 == 0:
    w1 = sess.run(w)[0]
    b1 = sess.run(b)[0]

    print(step,'\t', 1, '\t', b1)
```

```
plt.plot(vb[:40],'.-')
plt.xlabel('epoch')
plt.ylabel('b1')
```

```
nb_epoch = 1001
vw=[]

for step in range(nb_epoch):
  sess.run(train)
  w1 = sess.run(w)
  vw.append(w1)

  if step % 50 == 0:
    w1 = sess.run(w)[0]
    b1 = sess.run(b)[0]

    print(step,'\t', 1, '\t', w1)
```

```
plt.plot(vw[:50],'.-')
plt.xlabel('epoch')
plt.ylabel('w')
```
